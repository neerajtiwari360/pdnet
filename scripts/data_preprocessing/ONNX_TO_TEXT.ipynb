```python
import onnx

def convert_onnx_to_text(onnx_path, text_path):
    # Load the ONNX model
    model = onnx.load(onnx_path)

    # Convert the model to a string representation
    model_str = str(model)

    # Write the string representation to a text file
    with open(text_path, 'w') as f:
        f.write(model_str)

if __name__ == "__main__":
    # Path to your ONNX file
    onnx_path = '/content/drive/MyDrive/PIDNet/github/pdnet/model/onnx_model/model.onnx'

    # Path to the output text file
    text_path = '/content/drive/MyDrive/PIDNet/github/pdnet/model/onnx_model/model.txt'

    # Convert the ONNX model to a text file
    convert_onnx_to_text(onnx_path, text_path)
    print(f"ONNX model has been converted to text and saved to {text_path}")
```
